> ##### D6: The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:

> - If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
> - If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# The first step is to create a new variable to base my spending flag on. 

ords_prods_new_var['avg_spent'] = ords_prods_new_var.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Checking the results of the query / checking the new column. 

ords_prods_new_var[['user_id', 'order_number', 'prices', 'avg_spent']].tail(100)
```

> Now creating the flags according to brief: 

```python
ords_prods_new_var.loc[ords_prods_new_var['avg_spent'] < 10, 'spender_flag'] = 'Low spender'
```

```python
ords_prods_new_var.loc[ords_prods_new_var['avg_spent'] >= 10, 'spender_flag'] = 'High spender'
```

```python
# Checking the results

ords_prods_new_var[['user_id', 'order_number', 'prices', 'avg_spent', 'spender_flag']].head(100)
```

> ##### D7: In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
> - If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
> - If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
> - If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# The first step is to create a new variable to base my order frequency flag on. 

ords_prods_new_var['median_order_freq'] = ords_prods_new_var.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Checking the results of the query / checking the new column. 

ords_prods_new_var[['user_id', 'order_number', 'days_since_prior_order', 'median_order_freq']].head(100)
```

> Now creating the flags according to brief: 

```python
ords_prods_new_var.loc[ords_prods_new_var['median_order_freq'] > 20, 'order_freq_flag'] = 'Non-frequent customer'
```

```python
ords_prods_new_var.loc[(ords_prods_new_var['median_order_freq'] > 10) & (ords_prods_new_var['median_order_freq'] <= 20), 'order_freq_flag'] = 'Regular customer'
```

```python
ords_prods_new_var.loc[ords_prods_new_var['median_order_freq'] <= 10, 'order_freq_flag'] = 'Frequent customer'
```

```python
# Checking the results

ords_prods_new_var[['user_id', 'order_number', 'days_since_prior_order', 'median_order_freq', 'order_freq_flag']].head(100)
```
